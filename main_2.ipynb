## Data Visualization with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

import geopandas as gpd
from shapely.geometry import Point
```

### Basic plotting (line plots, scatter plots, histograms)

```python
df = pd.read_csv('dataset/data.csv')

# Scatter Plot (Age vs. Salary)
plt.scatter(df['Age'], df['Salary'])
plt.title("Age vs. Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# Histogram (Distribution of Age)
plt.hist(df['Age'], bins=2)  # Bins define the number of intervals
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bar plot Salary per City
salary_by_city = df.groupby('City')['Salary'].mean()
plt.bar(salary_by_city.index, salary_by_city.values)
plt.title("Average Salary by City")
plt.xlabel("City")
plt.ylabel("Average Salary")
plt.show()
```

### Creating maps (basemaps, geographic projections)

```python
# Example data (replace with your data)
data = {'City': ['New York', 'London', 'Paris', 'Tokyo'],
        'Latitude': [40.7128, 51.5074, 48.8566, 35.6895],
        'Longitude': [-74.0060, -0.1278, 2.3522, 139.6917]}
map_df = pd.DataFrame(data)

# Create GeoDataFrame
geometry = [Point(xy) for xy in zip(map_df['Longitude'], map_df['Latitude'])]
gdf = gpd.GeoDataFrame(map_df, geometry=geometry, crs="EPSG:4326")  # WGS 84

# Load a world map (or a more specific shapefile)
url = "https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip"
world = gpd.read_file(url)

# Plot the world map and your points
fig, ax = plt.subplots(figsize=(10, 6))
world.plot(ax=ax, alpha=0.5, color='gray') #basemap
gdf.plot(ax=ax, marker='o', color='red', markersize=50) #data plotted on basemap
plt.title("Cities on World Map")
plt.show()
```

### Customizing plots (colors, markers, labels)

```python
# Scatter Plot with Customizations
plt.scatter(df['Age'], df['Salary'], color='green', marker='x', s=100) #s is marker size
plt.title("Age vs. Salary (Customized)")
plt.xlabel("Age (Years)", fontsize=12)
plt.ylabel("Salary (USD)", fontsize=12)
plt.grid(True) #add grid lines.
plt.show()
```

```python
# Histogram with Customizations
plt.hist(df['Age'], bins=3, color='skyblue', edgecolor='black')
plt.title("Age Distribution (Customized)")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(range(min(df['Age']), max(df['Age'])+1, 5)) #Set specific x ticks.
plt.show()
```

```python
#Bar plot customizations
salary_by_city = df.groupby('City')['Salary'].mean()
plt.bar(salary_by_city.index, salary_by_city.values, color = ['red', 'blue', 'green', 'orange'])
plt.title("Average Salary by City")
plt.xlabel("City")
plt.ylabel("Average Salary")
plt.xticks(rotation = 45, ha = 'right') #rotate x axis labels.
plt.tight_layout() #prevent labels from being cut off.
plt.show()
```

Key Customization Elements:
- Colors: Use color names (e.g., 'red', 'blue', 'green'), hex codes, or RGB tuples.
- Markers: Change marker shapes (e.g., 'o', 'x', '^', 's').
- Line Styles: Modify line styles (e.g., '-', '--', ':').
- Labels: Add titles, axis labels, and legends.
- Grid: Add grid lines for better readability.
- Font Sizes: Adjust font sizes of labels and titles.
- Figure Size: Adjust the size of the plot.
- Saving plots: `plt.savefig('myplot.png')`

Matplotlib offers extensive customization options, allowing you to create informative and visually appealing plots. GeoPandas extends these capabilities for geographical data visualization.
